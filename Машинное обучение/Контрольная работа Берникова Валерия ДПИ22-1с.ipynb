5. (2 балла) Обучите полиномиальную модель классификации. Рассчитайте метрики accuracy и выведите таблицу классификации. Попробуйте разные степени полинома и выберите ту, которая работает лучше.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
best_accuracy = 0
best_degree = 1
```

```python
for degree in range(1, 6):  # Попробуем степени от 1 до 5
    model = make_pipeline(PolynomialFeatures(degree), LogisticRegression())
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy for degree {degree}: {accuracy}")
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_degree = degree

print(f"\nBest degree: {best_degree}")
```

```python
best_model = make_pipeline(PolynomialFeatures(best_degree), LogisticRegression())
best_model.fit(X_train, y_train)
```

```python
y_pred_best = best_model.predict(X_test)
```

```python
accuracy_poly = accuracy_score(y_test, y_pred_best)
accuracy_poly
```

```python
# Вывод таблицы классификации для лучшей модели
print("\nТаблица классификации для лучшей модели:")
print(classification_report(y_test, y_pred_best))
```

6. (2 балла) Обучите модель классификации по методу опорных векторов. Рассчитайте метрики accuracy и выведите таблицу классификации. Попробуйте разные ядерные функции и выберите ту, которая работает лучше.

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)

best_accuracy = 0
best_kernel = ""

kernels = ['linear', 'poly', 'rbf', 'sigmoid']
```

```python
for kernel in kernels:
    model = SVC(kernel=kernel)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy for kernel {kernel}: {accuracy}")
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_kernel = kernel
```

```python
best_model = SVC(kernel=best_kernel)
best_model.fit(X_train, y_train)
```

```python
y_pred_best = best_model.predict(X_test)
```

```python
accuracy_svm = accuracy_score(y_test, y_pred_best)
accuracy_svm
```

```python
# Вывод таблицы классификации для лучшей модели
print("\nТаблица классификации для лучшей модели:")
print(classification_report(y_test, y_pred_best))
```
